```python
import os
import shutil

# Folder path
source_dir = '/content/Asirra: cat vs dogs'
os.makedirs(f'{source_dir}/cats', exist_ok=True)
os.makedirs(f'{source_dir}/dogs', exist_ok=True)

# Files ko move karein
for filename in os.listdir(source_dir):
    if filename.startswith('cat') and filename.endswith('.jpg'):
        shutil.move(os.path.join(source_dir, filename), os.path.join(source_dir, 'cats', filename))
    elif filename.startswith('dog') and filename.endswith('.jpg'):
        shutil.move(os.path.join(source_dir, filename), os.path.join(source_dir, 'dogs', filename))

print("Files move ho gayi hain!")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Folder ka path manually copy kar lein (left sidebar mein folder par right-click karke 'Copy path' karein)
train_ds = tf.keras.utils.image_dataset_from_directory(
    '/content/Asirra: cat vs dogs',  # Agar ye main folder hai
    image_size=(150, 150),
    batch_size=32,
    label_mode='binary'
)

```

```python
import os
print(os.listdir('/content/Asirra: cat vs dogs'))
```

```python
from tensorflow.keras import layers, models

# 1. Model Architecture Define Karein
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(150, 150, 3)),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid') # Binary classification
])

# 2. Compile karein
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 3. Train karein
model.fit(train_ds, epochs=10)

# 4. Model Save karein
model.save('cat_dog_model.keras')
```
